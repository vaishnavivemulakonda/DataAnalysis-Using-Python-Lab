```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_value = 15
hours = 8

# Simulate calls per hour (Poisson distribution)
calls_per_hour = np.random.poisson(lam=lambda_value, size=hours)

# Plot histogram of calls per hour
plt.figure(figsize=(8,5))
plt.hist(calls_per_hour, bins=range(min(calls_per_hour), max(calls_per_hour)+2), alpha=0.75, color='skyblue', edgecolor='black')
plt.title('Histogram of Hourly Call Counts (λ=15)')
plt.xlabel('Number of Calls')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# Output mean value of calls per hour
mean_calls = np.mean(calls_per_hour)
print(f'Mean Calls Per Hour: {mean_calls}')
print(f'Hourly Calls: {calls_per_hour.tolist()}')
```

```python
# Different λ values and sample size
lambda_values = [5, 10, 15, 25, 40]
size = 1000

fig, axes = plt.subplots(len(lambda_values), 1, figsize=(8, 12))

for i, lam in enumerate(lambda_values):
    samples = np.random.poisson(lam=lam, size=size)
    axes[i].hist(samples, bins=range(min(samples), max(samples)+2), alpha=0.75, color='lightgreen', edgecolor='black')
    axes[i].set_title(f'Poisson Distribution (λ = {lam})')
    axes[i].set_xlabel('Number of Calls')
    axes[i].set_ylabel('Frequency')
    axes[i].grid(axis='y')

plt.tight_layout()
plt.show()
```
