```python
import numpy as np
import matplotlib.pyplot as plt

# Synthetic dataset for demonstration
np.random.seed(42)
num_legit = 30
num_spam = 30

offer_legit = np.random.poisson(1, num_legit)
buy_legit = np.random.poisson(2, num_legit)
offer_spam = np.random.poisson(5, num_spam)
buy_spam = np.random.poisson(6, num_spam)

plt.figure(figsize=(8,6))
plt.scatter(offer_legit, buy_legit, c='green', marker='^', label='Legitimate (label=0)')
plt.scatter(offer_spam, buy_spam, c='orange', marker='s', label='Spam (label=1)')
plt.xlabel('Count of the word "offer"')
plt.ylabel('Count of the word "buy"')
plt.title('Distribution of Offer & Buy Word Usage in Emails')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Synthetic dataset for demonstration
hours_studied = np.random.uniform(2, 10, 50)
test_scores = 50 + hours_studied * 5 + np.random.normal(0, 5, 50)
high_score = test_scores > 85

plt.figure(figsize=(8,6))
plt.scatter(hours_studied[~high_score], test_scores[~high_score], color='blue', label='Score â‰¤ 85', marker='o')
plt.scatter(hours_studied[high_score], test_scores[high_score], color='red', label='Score > 85', marker='*')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Study Hours vs Test Scores')
plt.legend()
plt.grid(True)
plt.show()
```
