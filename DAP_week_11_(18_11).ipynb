```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)

# -----------------------------
# A. BINOMIAL ANALYSIS
# -----------------------------
n = 60
p = 0.78
k = np.arange(0, n+1)

# 1. P(X = 50)
px_eq_50 = stats.binom.pmf(50, n, p)
print("P(X = 50) =", px_eq_50)

# 2. P(X â‰¤ 45)
px_le_45 = stats.binom.cdf(45, n, p)
print("P(X â‰¤ 45) =", px_le_45)

# 3. 95th percentile
q95 = stats.binom.ppf(0.95, n, p)
print("95th percentile =", q95)

# PMF + CDF values
binom_pmf = stats.binom.pmf(k, n, p)
binom_cdf = stats.binom.cdf(k, n, p)

# 4. Simulation (5000 trials)
sim_trials = stats.binom.rvs(n, p, size=5000)

# Histogram
plt.figure(figsize=(8,4))
plt.hist(sim_trials, bins=range(0, n+2), rwidth=0.8)
plt.title("Histogram: 5000 Simulated Binomial Trials (n=60, p=0.78)")
plt.xlabel("Number of immune participants")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# 5. Binomial PMF (FIXED stem)
plt.figure(figsize=(8,4))
plt.stem(k, binom_pmf)
plt.title("Binomial PMF (n=60, p=0.78)")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.tight_layout()
plt.show()

# 6. Binomial CDF
plt.figure(figsize=(8,4))
plt.step(k, binom_cdf, where='post')
plt.title("Binomial CDF (n=60, p=0.78)")
plt.xlabel("k")
plt.ylabel("P(X â‰¤ k)")
plt.tight_layout()
plt.show()

# Visualize-style PMF
plt.figure(figsize=(8,4))
plt.bar(k, binom_pmf)
plt.title("Binomial PMF (Visualize-style)")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.tight_layout()
plt.show()

# Visualize-style CDF
plt.figure(figsize=(8,4))
plt.plot(k, binom_cdf, marker='o')
plt.title("Binomial CDF (Visualize-style)")
plt.xlabel("k")
plt.ylabel("P(X â‰¤ k)")
plt.tight_layout()
plt.show()

# -----------------------------
# B. POISSON ANALYSIS
# -----------------------------
lam = 5
x = np.arange(0, 16)

# 1. P(X = 8)
p_eq_8 = stats.poisson.pmf(8, lam)
print("P(X = 8) =", p_eq_8)

# 2. P(X â‰¤ 5)
p_le_5 = stats.poisson.cdf(5, lam)
print("P(X â‰¤ 5) =", p_le_5)

# 3. 90th percentile
q90 = stats.poisson.ppf(0.90, lam)
print("90th percentile =", q90)

# PMF + CDF
poisson_pmf = stats.poisson.pmf(x, lam)
poisson_cdf = stats.poisson.cdf(x, lam)

# 4. Simulate 30 days
poisson_30 = stats.poisson.rvs(lam, size=30)

plt.figure(figsize=(9,3))
plt.plot(np.arange(1,31), poisson_30, marker='o')
plt.title("30-Day Simulation of Poisson Side-Effect Reports (Î»=5)")
plt.xlabel("Day")
plt.ylabel("Reports")
plt.grid(True)
plt.tight_layout()
plt.show()

# 5. Poisson PMF (FIXED stem)
plt.figure(figsize=(8,4))
plt.stem(x, poisson_pmf)
plt.title("Poisson PMF (Î»=5)")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.tight_layout()
plt.show()

# Visualize-style PMF
plt.figure(figsize=(8,4))
plt.bar(x, poisson_pmf)
plt.title("Poisson PMF (Visualize-style)")
plt.xlabel("k")
plt.ylabel("P(X=k)")
plt.tight_layout()
plt.show()

# Visualize-style CDF
plt.figure(figsize=(8,4))
plt.plot(x, poisson_cdf, marker='o')
plt.title("Poisson CDF (Visualize-style)")
plt.xlabel("k")
plt.ylabel("P(X â‰¤ k)")
plt.tight_layout()
plt.show()
```
