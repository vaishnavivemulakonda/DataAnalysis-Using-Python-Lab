```python
# ----------------------------------------------------------
# BINOMIAL DISTRIBUTION  SPORTS EXAMPLE
# Basketball free-throw analysis using simulation and theory
# ----------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt
from math import comb, sqrt

# ----------------------------
# 1. Parameters and simulation
# ----------------------------
p = 0.7        # probability of successful shot
n = 10         # number of free throws per game
games = 100    # number of games (simulations)

# Simulate number of successful free throws in 100 games
results = np.random.binomial(n, p, size=games)

# ----------------------------
# 2. Theoretical Distribution
# ----------------------------
k_values = np.arange(0, n + 1)
theoretical_pmf = [comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in k_values]

# ----------------------------
# Plot Simulation vs Theoretical Distribution
# ----------------------------
plt.figure(figsize=(8, 5))
plt.hist(results, bins=np.arange(-0.5, n + 1.5, 1), alpha=0.6, label='Simulation (100 games)', density=True)
plt.plot(k_values, theoretical_pmf, 'ro-', label='Theoretical Binomial PMF')
plt.title("Basketball Free-Throw Simulation (p=0.7, n=10)")
plt.xlabel("Number of Successful Free Throws")
plt.ylabel("Probability")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# ----------------------------
# 3. Distribution Characteristics
# ----------------------------
mean = n * p
variance = n * p * (1 - p)
std_dev = sqrt(variance)
mode = int((n + 1) * p)  # most likely number of successes

print("=== Distribution Characteristics ===")
print(f"Mean (expected successes): {mean:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Most Likely Number of Successful Free Throws: {mode}\n")

# ----------------------------
# 4. Exact Probability Calculations
# ----------------------------
# Probability of exactly 7 successes
P_X_eq_7 = comb(n, 7) * (p ** 7) * ((1 - p) ** (n - 7))

# Probability of at least 8 successes (X >= 8)
P_X_ge_8 = sum(comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in range(8, n + 1))

# Probability of fewer than 5 successes (X < 5)
P_X_lt_5 = sum(comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in range(0, 5))

print("=== Exact Probability Questions ===")
print(f"P(X = 7)  = {P_X_eq_7:.4f}  The player has about a {P_X_eq_7*100:.1f}% chance of scoring exactly 7.")
print(f"P(X  8)  = {P_X_ge_8:.4f}  Probability of scoring at least 8 baskets.")
print(f"P(X < 5)  = {P_X_lt_5:.4f}  Probability of scoring fewer than 5 baskets.\n")

print("Interpretation:")
print("- The high probability for X8 shows the player is quite consistent and reliable.")
print("- A small probability for X<5 indicates poor games are rare.\n")

# ----------------------------
# 5. What-if Scenario (p = 0.5)
# -
```
